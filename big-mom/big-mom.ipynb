```python
import pandas as pd
```

```python
def get_data_frame(csv):
    df = pd.read_csv(csv)
    hy = df['收盘价'][120:]
    hyf = pd.DataFrame(hy)
    hyf.columns = ['半年前收盘价']
    hyf = hyf.reset_index().drop(columns=['index'])
    return pd.merge(df, hyf, left_index=True, right_index=True)
```

```python
with open('399300.csv', encoding='gbk') as f:
    df = get_data_frame(f)
```

```python
df[df['收盘价'] / df['半年前收盘价'] > 1.5]
```
